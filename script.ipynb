```python
import pandas as pd
data_files = [
    "ap_2010.csv",
    "Class_Size.csv",
    "demographics.csv",
    "graduation.csv",
    "hs_directory.csv",
    "sat_results.csv"
]
data = {}
for csvs in data_files:
    files = pd.read_csv("{0}".format(csvs))
    key_name = csvs.replace(".csv", "")
    data[key_name] = files
```

```python
all_survey= pd.read_csv('survey_all.txt', delimiter="\t", encoding="windows-1252")
d75_survey = pd.read_csv('survey_d75.txt', delimiter="\t", encoding="windows-1252")
survey = pd.concat([all_survey, d75_survey], axis=0, sort=True)
survey.head()
```

```python
survey["DBN"] = survey["dbn"]
survey_fields = ["DBN", "rr_s", "rr_t", "rr_p", "N_s", "N_t", "N_p", "saf_p_11", "com_p_11", "eng_p_11", "aca_p_11", "saf_t_11", "com_t_11", "eng_t_11", "aca_t_11", "saf_s_11", "com_s_11", "eng_s_11", "aca_s_11", "saf_tot_11", "com_tot_11", "eng_tot_11", "aca_tot_11"]
survey = survey.loc[:, survey_fields]
data["survey"] = survey
survey.head()
```

```python
data["hs_directory"]["DBN"] = data["hs_directory"]["dbn"]

def psd_csd(num):
    return str(num).zfill(2)

data["Class_Size"]["padded_csd"] = data["Class_Size"]["CSD"].apply(psd_csd)
data["Class_Size"]["DBN"] = data["Class_Size"]["padded_csd"] + data["Class_Size"]["SCHOOL CODE"]
data["Class_Size"].head()
```

```python
data["Class_Size"].rename(columns = lambda x: x.strip())
```

```python
data["sat_results"]["SAT Math Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Math Avg. Score"], errors='coerce')
data["sat_results"]["SAT Critical Reading Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Critical Reading Avg. Score"], errors='coerce')
data["sat_results"]["SAT Writing Avg. Score"] = pd.to_numeric(data["sat_results"]["SAT Writing Avg. Score"], errors='coerce')
data["sat_results"]["sat_score"] = data["sat_results"]["SAT Math Avg. Score"] + data["sat_results"]["SAT Critical Reading Avg. Score"] + data["sat_results"]["SAT Writing Avg. Score"]
data["sat_results"]["sat_score"].head()
```

```python
import re
def extract_coordinate(string):
    pattern = '\(.+\)'
    coordiate = re.findall(pattern, string)
    latt = coordiate[0].split(",")[0].replace("(", "")
    return latt
data["hs_directory"]["lat"] = data["hs_directory"]["Location 1"].apply(extract_coordinate)
data["hs_directory"].head()
```

```python
import re
def conver_longt(string):
    pattern = "\(.+\)"
    coords = re.findall(pattern, string)
    lon = coords[0].split(",")[1].strip().replace(")","")
    return lon

data["hs_directory"]["lon"] = data["hs_directory"]["Location 1"].apply(conver_longt)
data["hs_directory"]["lat"] = pd.to_numeric(data["hs_directory"]["lat"], errors='coerce')
data["hs_directory"]["lon"] = pd.to_numeric(data["hs_directory"]["lon"], errors='coerce')

data["hs_directory"].head()
```
